# Dataset Cleaning

```python
cols = df.columns

for col in cols:
  print(f'{col}: {df[col].dtypes}')
```

```python
convert_cols_to_float = ['IPM', 'Persentase Penduduk Miskin', 'Laju Pertumbuhan PDRB atas dasar konstan 2010']

for col in convert_cols_to_float:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .str.extract(r'(\d+\.\d+|\d+)')
        .astype(float)
    )
```

```python
remove_rupiah = ['Pagu TKDD', 'Realisasi TKDD']

for col in remove_rupiah:
    # Remove 'Rp ', '.', and ',00'
    df[col] = df[col].astype(str).str.replace('Rp ', '', regex=False).str.replace('.', '', regex=False).str.replace(',00', '', regex=False)
    # Convert to integer, coercing errors to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

```python
df['Jumlah Penduduk'] = df['Jumlah Penduduk'].astype(str).str.replace('.', '', regex=False)
df['Jumlah Penduduk'] = pd.to_numeric(df['Jumlah Penduduk'], errors='coerce').astype('Int64')
```

```python
columns_to_convert_float = ['Anggaran APBN per kapita']

for col in columns_to_convert_float:
    df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['Persentase Penduduk Miskin'] = df['Persentase Penduduk Miskin'].astype(str).str.replace(',', '.', regex=False)
df['Persentase Penduduk Miskin'] = pd.to_numeric(df['Persentase Penduduk Miskin'], errors='coerce')
```

```python
pdrb_berlaku_col = 'Produk Domestik Regional Bruto (PDRB) Atas Dasar Harga Berlaku'
df[pdrb_berlaku_col] = df[pdrb_berlaku_col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
df[pdrb_berlaku_col] = pd.to_numeric(df[pdrb_berlaku_col], errors='coerce')
```

```python
pdrb_per_kapita_col = 'Produk Domestik Regional Bruto per Kapita HB'
df[pdrb_per_kapita_col] = df[pdrb_per_kapita_col].astype(str).str.replace('.', '', regex=False)
df[pdrb_per_kapita_col] = pd.to_numeric(df[pdrb_per_kapita_col], errors='coerce').astype('Int64')
```

```python
cols_to_impute_median = ['Pagu TKDD', 'Realisasi TKDD']

provinsi = ['SULAWESI BARAT', 'SULAWESI SELATAN', 'SULAWESI TENGAH', 'SULAWESI TENGGARA']

for col in cols_to_impute_median:
    # Calculate the median only for the specified provinces
    median_value = df[df['Provinsi'].isin(provinsi)][col].median()
    # Fill missing values in the specified provinces with their median
    df[col] = df[col].fillna(median_value)
```

```python

import numpy as np

def calculate_population(row):
    pdrb = row['Produk Domestik Regional Bruto (PDRB) Atas Dasar Harga Berlaku']
    pdrb_per_capita = row['Produk Domestik Regional Bruto per Kapita HB']
    # Cek apakah data valid dan tidak terjadi pembagian dengan nol
    if not pd.isnull(pdrb) and not pd.isnull(pdrb_per_capita) and pdrb_per_capita != 0:
        population = (pdrb * 1000000000) / pdrb_per_capita
        return int(population)  # Konversi ke integer
    return np.nan  # Kembalikan NaN jika perhitungan tidak valid


df['Jumlah Penduduk'] = df.apply(calculate_population, axis=1)


columns_to_show = [
    'Produk Domestik Regional Bruto (PDRB) Atas Dasar Harga Berlaku',
    'Produk Domestik Regional Bruto per Kapita HB',
    'Jumlah Penduduk',

]

# Tampilkan hasil
print("Baris dengan Jumlah Penduduk null dan perhitungan berdasarkan rumus:")
df
```

```python
def calculate_population(row):
    realisasi = row['Realisasi TKDD']
    jumlah = row['Jumlah Penduduk']
    if not pd.isnull(realisasi) and not pd.isnull(jumlah):
        hasil= realisasi / jumlah
        return int(hasil)  # Konversi ke integer
    return np.nan


df['Anggaran APBN per kapita'] = df.apply(calculate_population, axis=1)

# Tampilkan hasil
print("Baris dengan Jumlah Penduduk null dan perhitungan berdasarkan rumus:")
df
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Create a new DataFrame by dropping rows with missing values in the selected columns
df_cleaned_ipm = df.dropna(subset=['IPM', 'Produk Domestik Regional Bruto per Kapita HB', 'Produk Domestik Regional Bruto (PDRB) Atas Dasar Harga Berlaku', 'Anggaran APBN per kapita'])

# Define the features (independent variables) X
X = df_cleaned_ipm[['Produk Domestik Regional Bruto per Kapita HB', 'Produk Domestik Regional Bruto (PDRB) Atas Dasar Harga Berlaku', 'Anggaran APBN per kapita']]

# Define the target variable (dependent variable) y
y = df_cleaned_ipm['IPM']

# Instantiate a LinearRegression model
ipm_model = LinearRegression()

# Fit the regression model
ipm_model.fit(X, y)
```

```python
# Identify rows with missing 'IPM' values
missing_ipm_rows = df[df['IPM'].isnull()]

# Select the feature columns for these rows
X_missing_ipm = missing_ipm_rows[['Produk Domestik Regional Bruto per Kapita HB', 'Produk Domestik Regional Bruto (PDRB) Atas Dasar Harga Berlaku', 'Anggaran APBN per kapita']]

# Predict the missing 'IPM' values
predicted_ipm = ipm_model.predict(X_missing_ipm)

# Fill the missing 'IPM' values in the original DataFrame with the predictions
df.loc[df['IPM'].isnull(), 'IPM'] = predicted_ipm

# Display the updated rows to verify the imputation
display(df.loc[missing_ipm_rows.index][['Provinsi', 'IPM']])
```

```python
from sklearn.linear_model import LinearRegression

# Create a new DataFrame by dropping rows with missing values in the selected columns
df_cleaned_persentase = df.dropna(subset=['Persentase Penduduk Miskin', 'Produk Domestik Regional Bruto per Kapita HB', 'IPM', 'Realisasi TKDD', 'Pagu TKDD'])

# Define the features (independent variables) X
X_persentase = df_cleaned_persentase[['Produk Domestik Regional Bruto per Kapita HB', 'IPM', 'Realisasi TKDD', 'Pagu TKDD']]

# Define the target variable (dependent variable) y
y_persentase = df_cleaned_persentase['Persentase Penduduk Miskin']

# Instantiate a LinearRegression model
persentase_model = LinearRegression()

# Fit the regression model
persentase_model.fit(X_persentase, y_persentase)
```

## Univariate EDA

```python
import matplotlib.pyplot as plt
def perform_univariate_eda(df):
    for col in df.columns:
        print(f"\n--- Univariate Analysis for: {col} ---")
        print(f"Data Type: {df[col].dtype}")
        print(f"Number of unique values: {df[col].nunique()}")
        print(f"Number of missing values: {df[col].isna().sum()}")

        if df[col].dtype in ['float64', 'Int64']: # Numerical columns
            print("\nSummary Statistics:")
            print(df[col].describe())

            # Histograms for numerical data
            plt.figure(figsize=(8, 5))
            df[col].hist(bins=30)
            plt.title(f'Histogram of {col}')
            plt.xlabel(col)
            plt.ylabel('Frequency')
            plt.show()

            # Box plots for numerical data (to visualize outliers)
            plt.figure(figsize=(8, 5))
            df.boxplot(column=col)
            plt.title(f'Box Plot of {col}')
            plt.ylabel(col)
            plt.show()

        elif df[col].dtype == 'object': # Categorical columns
            print("\nValue Counts:")
            print(df[col].value_counts())

            # Bar plots for categorical data
            plt.figure(figsize=(10, 6))
            df[col].value_counts().plot(kind='bar')
            plt.title(f'Bar Plot of {col}')
            plt.xlabel(col)
            plt.ylabel('Count')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()

        else:
            print("\nNo specific EDA visualization for this data type.")
perform_univariate_eda(df)
```

```python
from scipy.stats import pearsonr
import pandas as pd

def analyze_specific_correlations(df):
    """Analyze specific meaningful correlations"""
    correlations_of_interest = [
        ('IPM', 'Persentase Penduduk Miskin'),
        ('IPM', 'Produk Domestik Regional Bruto per Kapita HB'),
        ('Anggaran APBN per kapita', 'IPM'),
        ('Persentase Penduduk Miskin', 'Produk Domestik Regional Bruto per Kapita HB'),
        ('Laju Pertumbuhan PDRB atas dasar konstan 2010', 'IPM')
    ]

    results = []
    for var1, var2 in correlations_of_interest:
        corr_coef, p_value = pearsonr(df[var1], df[var2])
        results.append({
            'Variable 1': var1,
            'Variable 2': var2,
            'Correlation': corr_coef,
            'p-value': p_value,
            'Significant': 'Yes' if p_value < 0.05 else 'No',
            'Interpretation': interpret_correlation(corr_coef)
        })

    return pd.DataFrame(results)

def interpret_correlation(r):
    """Interpret correlation strength"""
    abs_r = abs(r)
    if abs_r >= 0.8:
        return "Very Strong"
    elif abs_r >= 0.6:
        return "Strong"
    elif abs_r >= 0.4:
        return "Moderate"
    elif abs_r >= 0.2:
        return "Weak"
    else:
        return "Very Weak"

specific_corr_df = analyze_specific_correlations(df)
print("\nSpecific Correlations of Interest:")
print(specific_corr_df)
```
